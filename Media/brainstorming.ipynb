## Silence mechanisms, memory extensions, and parameter explorations

Adapting the Silence implementation from Sohn 2025

```python
import numpy as np
totalN = 10000
supportersN = 0
for supportersN in range(0, totalN, 50):
    opponentsN = totalN - supportersN
    delta = (supportersN - opponentsN)/totalN
    # print(delta)
    for alpha in np.arange(0, 1, 0.05):
        for delta2 in np.arange(-1, 1, 0.05):
            expression = alpha*delta + (1 - alpha)*delta2
            if(expression < -1 or expression > 1):
                print(f"For alpha of {alpha} we have delta: {delta} and expression{expression}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
tau = 5
delta = np.arange(-1, 1, 0.001)
delta0 = -0.75

phi = 1/(1 + np.exp(tau*(delta - delta0)))

plt.figure()
plt.xlim([-1, 1])
plt.ylim([0, 1])
plt.xticks(np.arange(-1,1,0.1))
plt.xlabel("Fraction of Excess Supporters")
plt.ylabel("Confidence Threshold for Expression \n(|$o_{i}$(t)| must exceed this)")
plt.plot(delta, phi)
```
